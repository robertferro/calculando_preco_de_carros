## Criando uma lista com os endereços

```python
lista_pag = []
for i in range(1,1537):
    num_1 ='_Desde_'
    num_2 = i
    num = '{}{}'.format(num_1,num_2)
    lista_pag.append(num)


lista_pag_1 = lista_pag[::48]
lista_pag_1[0] = ''
```

```python
modelos = ['Agile','astra','Celta','Classic',
           'Cobalt','Corsa','Corsa-Classic','Corsa-Sedan','Cruze','Joy','Montana',
           'Onix','Prisma','S10','Spin','Tracker','Vectra',
           'Amarok','Crossfox','Fox','Fusca','Gol','Golf',
           'Jetta','Kombi','Parati','Passat','Polo','Polo-Sedan',
           'Saveiro','Spacefox','T-Cross','Up!','Virtus','Voyage',
           'Argo','Bravo','Cronos','Doblo','Ducato','Fiorino','Freemont',
           'Grand-Siena','Idea','Linea','Mobi','Palio','Palio-Weekend','Punto',
           'Siena','Stilo','Strada','Toro','Uno',
           'Ecosport','Escort','Fiesta','Fiesta-Sedan',
           'Focus','Focus-Sedan','Fusion','Ka','Ranger',
           'Creta','HB20','HB20S',
           'I30','IX35','Santa-Fe','Tucson',
           'Captur','Clio','Duster','Duster-Oroch',
           'Fluence','Kwid','Logan','Master','Sandero',
           'Corolla','etios','Hilux','RAV4','SW4','Yaris',
           'City','Civic','CR-V','Fit','HR-V','WR-V',
           '2008','206','207','207-Passion','208','3008','307','308','408',
           'Aircross','C3','C3-Picasso','C4','C4-Cactus',
           'C4-Lounge','C4-Pallas','C4-Picasso','Xsara-Picasso',
           'Cerato','Picanto','Soul','Sportage',
           'Frontier','Kicks','Livina','March','Sentra','Tiida','Versa',
           'ASX','L200','Lancer','Outlander','Pajero','Pajero-Full','Pajero-Sport','Pajero-TR4',

            ]

```

```python
URL = 'https://lista.mercadolivre.com.br/veiculos/carros-caminhonetes/carros'

paginas = lista_pag_1[0:2]
```

```python
def scrap(url):
    c = requests.get(url).content
    soup = BeautifulSoup(c)
    lista_preco_tags = soup.find_all('span', {'class':'price-tag-fraction'})
    lista_preco = [info.text for info in lista_preco_tags ]
    lista_preco = lista_preco[:48]
    lista_km_tags = soup.find_all('li', {'class':'ui-search-card-attributes__attribute'})
    lista_km = [info.text for info in lista_km_tags ]
    lista_km = lista_km[1::2]
    lista_ano_tags = soup.find_all('li', {'class':'ui-search-card-attributes__attribute'})
    lista_ano = [info.text for info in lista_ano_tags ]
    lista_ano = lista_ano[::2]
    lista_car_tags = soup.find_all('h2', {'class':'ui-search-item__title ui-search-item__group__element'})
    lista_car = [info.text for info in lista_car_tags ]
    df = convert_df(modelos,lista_preco, lista_km, lista_ano, lista_car )
    return df


def convert_df(modelos, lista_preco, lista_km, lista_ano, lista_car ):
    modelos_lista = [modelo]*len(lista_preco)
    data = {'modelo':modelos_lista,
            'preco':lista_preco,
            'quilometragem':lista_km,
            'ano':lista_ano,
            'infomacoes': lista_car,
            
            }
    df = pd.DataFrame(data)
    return df
```

```python
lista_dfs = []
for modelo in modelos:
    for pag in paginas:
        url ='https://lista.mercadolivre.com.br/veiculos/carros-caminhonetes/carros/{}/{}'.format(modelo,pag)
        print(url)
        df=scrap(url)
        lista_dfs.append(df)
        sleep(1)
```

 - Criando o df com o resultado da pesquisa

```python
df = pd.concat(lista_dfs)
```
